```python
table=pd.crosstab(wine1['country'], wine1['color'],  margins=True)
table
```

```python
table.info()
```

```python
table1=table.copy()
```

```python
table1['red_%'] = table1.red/table1.All
table1
```

```python
table2=table.copy()
```

```python
table2['white_%'] = table2.white/table2.All
table2
```

```python
goods=table.assign( Red=lambda x: (x['red'] / x['All']), White=lambda x: (x['white'] / x['All']))
goods
```

```python
r1 = goods[(goods.White >goods.Red)] [['White']]
r1
```

```python
r2 = goods[(goods.White <goods.Red)] [['Red']]
r2
```

```python
max_red=r2[r2['Red']==r2['Red'].max()]
max_white=r1[r1['White']==r1['White'].max()]
```
